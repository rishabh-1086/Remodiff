```python
from transformers import *
import librosa
import torch
import os
# import ffmpeg
import pydub
from moviepy.video.io.VideoFileClip import VideoFileClip
# from moviepy.editor import VideoFileClip
```

```python
# audio_path = "/project/msoleyma_1026/personality_detection/first_impressions_v2_dataset/audio_files"
# video_dir = "/project/msoleyma_1026/personality_detection/first_impressions_v2_dataset/training"
video_dir = "/project/msoleyma_1026/personality_detection/first_impressions_v2_dataset/testing/video_files"
audio_path = "/project/msoleyma_1026/personality_detection/first_impressions_v2_dataset/testing/audio_files"
```

```python
for train_fold in os.listdir(video_dir): 
    if train_fold.endswith(".zip") : 
        continue
    full_train_fold = os.path.join(video_dir, train_fold) 
    audio_fold = os.path.join(audio_path, train_fold) 
    if not os.path.isdir(audio_fold): 
        os.makedirs(audio_fold) 
    for fold in os.listdir(full_train_fold): 

        fold = os.path.join(full_train_fold, fold) 
        for vid in os.listdir(fold): 
            vid_full_path = os.path.join(fold, vid) 
            full_audio_file = os.path.join(audio_fold, vid[:-4]+".wav") 
            video = VideoFileClip(vid_full_path)
            # audio = pydub.AudioSegment.from_file(vid_full_path, format="mp4")
            video.audio.write_audiofile(full_audio_file)
```
